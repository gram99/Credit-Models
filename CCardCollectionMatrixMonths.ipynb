```python
import pandas as pd
import numpy as np

# 1. Define Vintage (Months) and Debt Amount Buckets
vintages = [f"{m} Mo" for m in range(6, 73, 6)]
amounts = ["$0-500", "$500-1k", "$1k-5k", "$5k-10k", "$10k-25k", "$25k+"]

# 2. Generate Synthetic Collection Probabilities
# Probabilities drop as vintage (months) and debt amount increase
data = np.zeros((len(vintages), len(amounts)))
for i in range(len(vintages)):
    for j in range(len(amounts)):
        # Prob decays 12% per 6mo vintage step and 5% per amount bucket step
        prob = 0.85 * (0.88 ** i) * (0.95 ** j)
        data[i, j] = max(prob, 0.02) # Floor at 2%

df = pd.DataFrame(data, index=vintages, columns=amounts)

# 3. Apply Gradient and Grid Lines
# 'Reds_r' reverses the map so high probabilities are white/light
styled_matrix = df.style.background_gradient(cmap='Reds_r', axis=None) \
    .format("{:.1%}") \
    .set_properties(**{
        'border': '1px solid black',
        'text-align': 'center',
        'padding': '10px'
    }) \
    .set_caption("Probability of Collection: Vintage vs. Debt Amount")

# Display in a Jupyter Notebook or save to HTML
styled_matrix
```
