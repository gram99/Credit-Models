```python
import pandas as pd
import numpy as np

# 1. Define dimensions for Vintage and Debt Amount
vintages = [f'Year {i}' for i in range(1, 8)]
amount_buckets = ['$0-1k', '$1k-5k', '$5k-10k', '$10k-25k', '$25k-50k', '$50k+']

# 2. Generate Synthetic Collection Probabilities
# Logic: Probabilities decay as vintage and amount increase
base_prob = 0.45
data = []

for v_idx in range(len(vintages)):
    row = []
    for a_idx in range(len(amount_buckets)):
        # Decay formula: older debt (v_idx) and larger debt (a_idx) reduce recovery
        prob = base_prob * (0.80 ** v_idx) * (0.95 ** a_idx)
        # Clip to ensure valid probability range
        row.append(max(min(prob, 1.0), 0.0))
    data.append(row)

df = pd.DataFrame(data, index=vintages, columns=amount_buckets)

# 3. Apply Gradient and Grid Lines
# Use 'Reds_r' (reversed) to make high probability white and low probability red
styled_matrix = df.style.background_gradient(cmap='Reds_r', axis=None) \
    .format("{:.2%}") \
    .set_properties(**{
        'border': '1px solid black',
        'text-align': 'center',
        'padding': '10px'
    }) \
    .set_caption("Collection Probability: Vintage vs. Debt Amount")

# Display the matrix (standard for Jupyter/Notebook environments)
styled_matrix
```
