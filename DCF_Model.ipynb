```python
import pandas as pd
from IPython.display import clear_output, display

def run_dcf_session():
    # 1. Internal State (The "Variables")
    state = {
        'fcf': 100.0,
        'growth': 0.05,
        'wacc': 0.10,
        't_growth': 0.02
    }

    while True:
        # --- CALCULATE ---
        forecast, pvs = [], []
        curr = state['fcf']
        for i in range(1, 6):
            curr *= (1 + state['growth'])
            forecast.append(curr)
            pvs.append(curr / (1 + state['wacc'])**i)
        
        tv = (forecast[-1] * (1 + state['t_growth'])) / (state['wacc'] - state['t_growth'])
        ev = sum(pvs) + (tv / (1 + state['wacc'])**5)
        
        # --- DISPLAY ---
        clear_output(wait=True)
        print("====================================================")
        print("           DCF SCENARIO MANAGER (STABLE)            ")
        print("====================================================")
        print(f" INPUTS: FCF: ${state['fcf']}M | Growth: {state['growth']*100}% ")
        print(f"         WACC: {state['wacc']*100}% | Term Growth: {state['t_growth']*100}%")
        print("-" * 52)
        print(f" ENTERPRISE VALUE: ${ev:,.2f} M")
        print(f" TERMINAL VALUE:   ${tv:,.2f} M")
        print("-" * 52)
        
        df = pd.DataFrame({
            "Year": [f"Year {i}" for i in range(1, 6)],
            "Forecasted FCF": [f"${x:,.2f}M" for x in forecast],
            "Present Value": [f"${x:,.2f}M" for x in pvs]
        }).set_index("Year")
        display(df)
        
        # --- BUSINESS RULE: INPUT TRIGGER ---
        print("\n[1] Change FCF  [2] Change Growth  [3] Change WACC  [4] Reset  [Q] Quit")
        cmd = input("Select an option to update variables & recalculate: ").lower().strip()

        if cmd == 'q': 
            break
        elif cmd == '4':
            state = {'fcf': 100.0, 'growth': 0.05, 'wacc': 0.10, 't_growth': 0.02}
        elif cmd in ['1', '2', '3']:
            try:
                new_val = float(input("Enter new value: "))
                if cmd == '1': state['fcf'] = new_val
                if cmd == '2': state['growth'] = new_val if new_val < 1 else new_val/100
                if cmd == '3': state['wacc'] = new_val if new_val < 1 else new_val/100
            except ValueError:
                input("Invalid input. Press Enter to continue...")

run_dcf_session()
```
