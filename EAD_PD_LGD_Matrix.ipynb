```python
import pandas as pd
import numpy as np

# 1. Define rating scales
pd_ratings = [str(i) for i in range(1, 15)]  # 1 (lowest) to 14 (highest)
lgd_grades = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']  # A (lowest) to H (highest)

# 2. Generate synthetic probabilities
# PD: Exponential growth from ~0.03% (safest) to ~45.0% (highest criticized)
pd_probs = np.geomspace(0.0003, 0.45, num=len(pd_ratings))

# LGD: Linear scale from 10% (Grade A) to 90% (Grade H)
lgd_probs = np.linspace(0.10, 0.90, num=len(lgd_grades))

# 3. Create Expected Loss (EL) Matrix: EL = PD * LGD
el_matrix = np.outer(pd_probs, lgd_probs)
df = pd.DataFrame(el_matrix, index=pd_ratings, columns=lgd_grades)

# 4. Apply Color Gradient and Table Lines
# This will render correctly in Jupyter Notebooks or as an HTML output
styled_df = df.style.background_gradient(cmap='Reds', axis=None) \
    .format("{:.2%}") \
    .set_properties(**{
        'border': '1px solid black',
        'text-align': 'center',
        'padding': '10px'
    }) \
    .set_caption("PD/LGD Expected Loss Heatmap (White-to-Red Gradient)")

# Display the styled table (in a notebook environment)
styled_df
```
