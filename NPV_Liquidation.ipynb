```python
import pandas as pd
from IPython.display import clear_output, display

def run_valuation_calculator():
    # 1. Initialize State
    data = {
        'discount_rate': 0.10,
        'liquidation_val': 500.0,
        'cash_flows': [100.0, 120.0, 150.0, 180.0, 200.0] # 5-Year Forecast
    }

    while True:
        # --- CALCULATIONS ---
        # Calculate NPV: Sum of CF / (1 + r)^t
        npv = sum([cf / (1 + data['discount_rate'])**(i+1) for i, cf in enumerate(data['cash_flows'])])
        
        # Determine Best Choice
        if npv > data['liquidation_val']:
            decision = "✅ CHOICE: OPERATE (NPV > Liquidation)"
            color_code = "Green"
        elif npv < data['liquidation_val']:
            decision = "⚠️ CHOICE: LIQUIDATE (Liquidation > NPV)"
            color_code = "Red"
        else:
            decision = "⚖️ CHOICE: INDIFFERENT (NPV == Liquidation)"
            color_code = "Blue"

        # --- DISPLAY RESULTS ---
        clear_output(wait=True)
        print("====================================================")
        print("       NPV vs. LIQUIDATION DECISION TOOL           ")
        print("====================================================")
        print(f" Discount Rate: {data['discount_rate']*100}%")
        print(f" Liquidation Value: ${data['liquidation_val']:,.2f}")
        print("-" * 52)
        print(f" CALCULATED NPV:   ${npv:,.2f}")
        print(f" {decision}")
        print("-" * 52)
        
        # Display Forecast Table
        df = pd.DataFrame({
            "Year": [f"Year {i+1}" for i in range(len(data['cash_flows']))],
            "Forecasted CF": [f"${x:,.2f}" for x in data['cash_flows']],
            "Present Value": [f"${x / (1 + data['discount_rate'])**(i+1):,.2f}" for i, x in enumerate(data['cash_flows'])]
        }).set_index("Year")
        display(df)

        # --- USER INPUT MENU ---
        print("\n[1] Change Discount Rate  [2] Change Liquidation Value")
        print("[3] Change Cash Flows     [R] Reset    [Q] Quit")
        
        choice = input("\nSelect option to update and recalculate: ").strip().lower()

        if choice == 'q':
            break
        elif choice == 'r':
            data['discount_rate'] = 0.10
            data['liquidation_val'] = 500.0
        elif choice == '1':
            try:
                val = float(input("Enter new Discount Rate (e.g. 0.12 for 12%): "))
                data['discount_rate'] = val if val < 1 else val/100
            except: pass
        elif choice == '2':
            try:
                data['liquidation_val'] = float(input("Enter new Liquidation Value ($): "))
            except: pass
        elif choice == '3':
            try:
                print("Enter 5 years of cash flows separated by commas (e.g. 100,110,120,130,140):")
                cf_input = input("New Cash Flows: ")
                data['cash_flows'] = [float(x.strip()) for x in cf_input.split(',')]
            except:
                print("Invalid format. Use numbers separated by commas.")
                input("Press Enter to continue...")

run_valuation_calculator()
```
