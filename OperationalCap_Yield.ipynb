```python
import pandas as pd
from IPython.display import clear_output, display

def run_capacity_model():
    # --- 1. Initial Business Logic State ---
    state = {
        'fte_count': 10,          # Number of Agents
        'hours_per_fte': 40,      # Weekly Hours
        'utilization': 0.80,      # % of time actually spent working accounts
        'avg_handle_time': 10,    # Minutes spent per account touch
        'total_cures': 500,       # Number of accounts successfully cured
        'recovered_value': 250000 # Total dollars recovered
    }

    while True:
        # --- 2. Calculations ---
        # Total Capacity in Minutes
        total_minutes = (state['fte_count'] * state['hours_per_fte'] * 60) * state['utilization']
        # Capacity in total possible "Touches"
        total_capacity_touches = total_minutes / state['avg_handle_time']
        
        # Yield Metrics
        cure_rate = (state['total_cures'] / total_capacity_touches) * 100 if total_capacity_touches > 0 else 0
        yield_per_touch = state['recovered_value'] / total_capacity_touches if total_capacity_touches > 0 else 0
        yield_per_fte = state['recovered_value'] / state['fte_count'] if state['fte_count'] > 0 else 0

        # --- 3. Display Interface ---
        clear_output(wait=True)
        print("====================================================")
        print("     OPERATIONAL CAPACITY & YIELD OPTIMIZER         ")
        print("====================================================")
        print(f" INPUTS: Agents: {state['fte_count']} | Weekly Hrs: {state['hours_per_fte']} | Util: {state['utilization']*100}%")
        print(f"         Handle Time: {state['avg_handle_time']} min | Cures: {state['total_cures']}")
        print("-" * 52)
        
        # Summary Metrics
        results_df = pd.DataFrame({
            "Metric": ["Total Capacity (Touches)", "Cure Rate %", "Yield per Touch ($)", "Yield per Agent ($)"],
            "Value": [
                f"{total_capacity_touches:,.0f} Touches",
                f"{cure_rate:.2f}%",
                f"${yield_per_touch:,.2f}",
                f"${yield_per_fte:,.2f}"
            ]
        }).set_index("Metric")
        
        display(results_df)
        print("-" * 52)

        # --- 4. Scenario Inputs ---
        print("\n Change Staffing (FTE/Hours/Util)")
        print(" Change Performance (Handle Time/Cures/Value)")
        print(" [R] Reset to Defaults    [Q] Quit")
        
        choice = input("\nSelect a category to update variables: ").strip().lower()

        if choice == 'q':
            break
        elif choice == 'r':
            state = {'fte_count': 10, 'hours_per_fte': 40, 'utilization': 0.80, 'avg_handle_time': 10, 'total_cures': 500, 'recovered_value': 250000}
        elif choice == '1':
            try:
                state['fte_count'] = int(input("Enter Number of Agents (FTE): "))
                state['hours_per_fte'] = float(input("Enter Hours per Week per Agent: "))
                util = float(input("Enter Utilization % (e.g. 0.85 for 85%): "))
                state['utilization'] = util if util < 1 else util/100
            except: pass
        elif choice == '2':
            try:
                state['avg_handle_time'] = float(input("Enter Average Handle Time (Minutes): "))
                state['total_cures'] = int(input("Enter Total Number of Cures: "))
                state['recovered_value'] = float(input("Enter Total Recovered Dollars ($): "))
            except: pass

run_capacity_model()
```
