```python
import pandas as pd
import numpy as np
from faker import Faker
import random

fake = Faker()
np.random.seed(42)

def generate_risk_portfolio(n=1000):
    data = []
    
    # PD Grades (1=Lowest Risk, 15=Highest Risk)
    # LGD Classes (A=Lowest Severity, J=Highest Severity)
    lgd_classes = list("ABCDEFGHIJ")
    
    for _ in range(n):
        # Generate Exposure at Default (EAD) 
        # Using lognormal to simulate realistic commercial loan sizes
        ead = np.random.lognormal(mean=11.5, sigma=1.2) 
        
        # Logic: Larger exposures often undergo stricter underwriting (lower PD)
        # but may have higher loss severity if they fail (higher LGD)
        if ead < 100000:
            pd_grade = random.randint(7, 15)   # Riskier small loans
            lgd_idx = random.randint(0, 3)     # Low severity (easier to recover)
        elif ead < 1000000:
            pd_grade = random.randint(3, 10)   # Mid-market stability
            lgd_idx = random.randint(2, 6)     
        else:
            pd_grade = random.randint(1, 6)    # High-quality large corporate
            lgd_idx = random.randint(5, 9)     # High severity (complex recovery)
            
        lgd_class = lgd_classes[lgd_idx]
        risk_rating = f"{pd_grade}{lgd_class}"
        
        data.append({
            'Borrower': fake.company(),
            'EAD': round(ead, 2),
            'PD_Grade': pd_grade,
            'LGD_Class': lgd_class,
            'Risk_Rating': risk_rating,
            'Industry': fake.bs().split(' ')[-1].capitalize()
        })
        
    return pd.DataFrame(data)

# Generate and display top risks
df = generate_risk_portfolio(1000)

print("### SYNTHETIC INHERENT RISK MODEL OUTPUT")
print("Top 10 Loans by Exposure:")
print(df.sort_values('EAD', ascending=False).head(10).to_string(index=False))
```
