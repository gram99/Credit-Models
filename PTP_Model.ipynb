```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_auc_score

# 1. Load your dataset (assuming a CSV with features and target 'paid')
df = pd.read_csv('collection_data.csv')

# 2. Simple Preprocessing
# Convert categorical features (e.g., employment_status) into numeric dummy variables
df = pd.get_dummies(df, drop_first=True)

# 3. Define Features (X) and Target (y)
X = df.drop('paid', axis=1)
y = df['paid']

# 4. Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 5. Build the Model (Random Forest is common for high accuracy)
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# 6. Predict Propensity Scores (probabilities)
# model.predict_proba returns [prob_0, prob_1]; we want prob_1 (propensity to pay)
propensity_scores = model.predict_proba(X_test)[:, 1]

# 7. Evaluate Performance
print(f"ROC-AUC Score: {roc_auc_score(y_test, propensity_scores):.4f}")
```
