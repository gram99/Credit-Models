```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

# 1. Configuration: Define your Debt Age unit here (e.g., "Days", "Months", or "Years")
TIME_UNIT = "Days" 

# 2. Generate Synthetic Dataset
np.random.seed(42)
n_samples = 20
names = ["James Smith", "Maria Garcia", "Robert Johnson", "David Miller", "Linda Davis",
         "Michael Rodriguez", "Mary Martinez", "William Hernandez", "Barbara Lopez", "Richard Gonzalez",
         "Joseph Wilson", "Susan Anderson", "Thomas Thomas", "Jessica Taylor", "Charles Moore",
         "Sarah Jackson", "Christopher Martin", "Karen Lee", "Daniel Perez", "Nancy Thompson"]

# Create features based on the chosen unit
debt_age = np.random.randint(15, 180, n_samples) # Example range for 'Days'
amount_owed = np.round(np.random.uniform(500, 15000, n_samples), 2)
past_behavior = np.random.uniform(0.1, 0.9, n_samples) # Score 0.1 to 0.9

# Create synthetic target logic
logit_sim = (3 * past_behavior) - (0.02 * debt_age) - (0.0001 * amount_owed) + 1
will_repay = (1 / (1 + np.exp(-logit_sim)) > 0.5).astype(int)

df = pd.DataFrame({
    'Borrower Name': names,
    f'Debt Age ({TIME_UNIT})': debt_age,
    'Amount Owed': amount_owed,
    'Past Behavior Score': past_behavior,
    'Will Repay': will_repay
})

# 3. Train Predictive Model
# Note: Logistic Regression typically performs better when features are scaled
X = df[[f'Debt Age ({TIME_UNIT})', 'Amount Owed', 'Past Behavior Score']]
y = df['Will Repay']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

model = LogisticRegression()
model.fit(X_scaled, y)

# 4. Predict Probabilities
# predict_proba returns [prob_class_0, prob_class_1]; we take index 1 for "Will Repay"
df['Prob. of Repayment'] = np.round(model.predict_proba(X_scaled)[:, 1], 4)

# 5. Final Predictive Report
report = df[['Borrower Name', f'Debt Age ({TIME_UNIT})', 'Amount Owed', 'Past Behavior Score', 'Prob. of Repayment']]
print(f"--- DEBTOR REPAYMENT PREDICTION (Unit: {TIME_UNIT}) ---")
print(report.to_string(index=False))

    
```
