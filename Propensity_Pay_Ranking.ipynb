```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

# 1. INPUT: User-identified spreadsheet
input_file = 'your_collections_data.xlsx' # Replace with your file
df = pd.read_excel(input_file)

# 2. DEFINE FEATURES: Update these to match your column names
# Historical data used to predict the 'Paid_Status' (1 for paid, 0 for unpaid)
features = ['days_overdue', 'total_balance', 'previous_on_time_payments']
target = 'Paid_Status'

# 3. TRAIN MODEL: Building the prediction logic
# This uses historical records to 'learn' payment patterns
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(df[features], df[target])

# 4. PREDICT & RANK: Calculate scores and sort them
# predict_proba returns [prob_unpaid, prob_paid]; we take the prob_paid index [:, 1]
df['propensity_score'] = model.predict_proba(df[features])[:, 1]

# Ranking lowest to highest propensity
ranked_df = df.sort_values(by='propensity_score', ascending=True)

# 5. OUTPUT: Save to a new ranked spreadsheet
ranked_df.to_excel('ranked_propensity_output.xlsx', index=False)
print("Ranked data saved to 'ranked_propensity_output.xlsx'")
```
