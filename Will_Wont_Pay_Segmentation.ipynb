```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier

# 1. Generate Synthetic Data
np.random.seed(42)
n = 20  # Reduced size for the sample report
data = {
    'Name': [f"Customer {i:03d}" for i in range(1, n + 1)],
    'Balance Due': np.round(np.random.uniform(500, 15000, n), 2),
    'Days Outstanding': np.random.randint(1, 120, n),
    'Credit Score': np.random.randint(300, 850, n),
    'Payment History Score': np.random.uniform(0, 1, n)
}
df = pd.DataFrame(data)

# 2. Define Propensity Logic
# Logic: Low credit score and high days outstanding indicate low propensity
df['Target'] = ((df['Days Outstanding'] / 120) - (df['Credit Score'] / 850) > 0.1).astype(int)

# 3. Train Predictive Model
X = df[['Balance Due', 'Days Outstanding', 'Credit Score', 'Payment History Score']]
y = df['Target']
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X, y)

# 4. Predict and Label
df['Propensity'] = model.predict(X)
df['Propensity Category'] = df['Propensity'].apply(
    lambda x: 'Low (Legal/Settlement)' if x == 1 else 'High (Reminder Only)'
)

# 5. GENERATE AND PRINT REPORT
report = df[['Name', 'Balance Due', 'Days Outstanding', 'Propensity Category']]
print("\n--- DEBTOR PROPENSITY REPORT ---")
print(report.to_string(index=False))
```
